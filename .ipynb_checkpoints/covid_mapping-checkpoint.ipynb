## COVID-19 Mapping

After importing all the relevant libraries and creating a dataframe containing the details of COVID-19 patients, we will figure out the longitude and latitude
1. Combine address fields and consolidate them as df.column
2. Perform geocoding on them to get long/lat
3. Make changes to the dataframe to store the values
4. Plot the heatmaps based on long/lat

Reference: https://python-googlegeocoder.readthedocs.io/en/latest/
Reference: https://towardsdatascience.com/easy-steps-to-plot-geographic-data-on-a-map-python-11217859a2db

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from googlegeocoder import GoogleGeocoder
import matplotlib as mpl
%matplotlib inline
geocoder = GoogleGeocoder("")
```

```python
df = pd.read_excel(r'enhanced_sur_covid_19_eng.xlsx')
df.head()
```

```python
df['Combined Address'] = df['Residence'] + ' ' + df['Area of Residence'] + ' Hong Kong'
```

```python
df['Combined Address']
```

```python
geo_code = []
for row_index,row in df.iterrows():
    try:
        search = geocoder.get(row['Combined Address'])
        geo_code.append(search[0].geometry.location)
    except ValueError:
        geo_code.append("Error")
```

```python
latitude = []
longitude = []
for g in geo_code:
    lat = float(g.lat)
    long = float(g.lng)
    latitude.append(lat)
    longitude.append(long)
```

```python
df['latitude'] = latitude
df['longitude'] = longitude
```

```python
df.to_csv('lat_long.csv')
df.head()
```

```python
BBox = (df.longitude.min(),   df.longitude.max(),      
         df.latitude.min(), df.latitude.max())
BBox
```

```python
BBoxOpen =(113.8527, 114.4247, 22.1810, 22.5576)
BBoxOpen
```

```python
dpi = 300
mpl.rcParams['figure.dpi']= dpi
```

```python
ruh_m = plt.imread('map.png')
fig, ax = plt.subplots(figsize = (9,8))
ax.scatter(df.longitude, df.latitude, zorder=1, alpha= 0.2, c='b', s=10)
ax.set_title('Plotting COVID-19 Data on Hong Kong Map')
ax.set_xlim(BBoxOpen[0],BBoxOpen[1])
ax.set_ylim(BBoxOpen[2],BBoxOpen[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```
